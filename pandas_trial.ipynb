```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import scipy.stats
```

```python
df = pd.read_csv("nifty-data.csv")
df.head()
```

```python
df.index = df['Date']
```

```python
df.head()
```

```python
df['Close'].plot(figsize = (12,6))
plt.show()
```

```python
df['Close'].plot(figsize=(12,6))
plt.xlim([0,50])
plt.show()
```

```python
df['Daily returns'] = df['Close'].pct_change()
df['Difference'] = df['Close'].diff()
df
```

```python
df['Difference'].plot(figsize=(12,6))
#df['Daily returns'].plot(figsize=(12,6))

plt.show()
```

```python
df['Daily returns'].hist(bins=20, figsize = (12,6))
plt.xlabel('Daily returns')
plt.ylabel('Frequency')
plt.show()
df['Daily returns'].describe()
```

```python
df.dropna(axis =0,inplace=True)
```

```python
_, bins, _=plt.hist(df['Daily returns'], 50 , density = 1 , alpha = 0.5)
mu, sigma = scipy.stats.norm.fit(df['Daily returns'])
bfl=scipy.stats.norm.pdf(bins, mu, sigma)
plt.plot(bins,bfl)
```
