```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
mu = 168
sigma = 5
sample = 250
np.random.seed(0)
height_f = np.random.normal(mu, sigma, sample).astype(int)
```

```python
print(np.ptp(height_f))
```

```python
mu = 176 #mean
sigma = 6 #stddev
sample = 250
np.random.seed(1)
height_m = np.random.normal(mu, sigma, sample).astype(int)
```

```python
gym = pd.DataFrame({'height_m':height_m, 'height_f':height_f})
```

```python
gym
```

```python
plt.figure(dpi=1200)
gym.plot(figsize=(10,5))
```

```python
gym.groupby('height_m').count()
gym.groupby('height_f').count()
gym.plot.hist(bins=20, alpha = 0.7)
```
